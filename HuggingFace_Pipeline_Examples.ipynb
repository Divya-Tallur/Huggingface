```python
# Named Entity Recognition
from transformers import pipeline

ner=pipeline("ner",grouped_entities=True)
text=" Hugging face Inc. is a company based in Newyork city"
entities=ner(text)
print(entities)
```

```python
# Question - Answering

qa=pipeline("question-answering")
context = "Divya is good girl. Her favourite food is Dosa. She is currently in Liverpool"
quetsion= "Where is Divya based out of?"
answer=qa(question=quetsion,context=context)
print(answer)
question2 = "What is her favourite food?"
answer2 = qa(question=question2, context=context)
print(answer2)
question3="What is her favourite sport?"
answer3=qa(question=question3,context=context)
print(answer3)
```

```python
# Summarization
summary=pipeline("summarization")
paragraph = "The cucumber is a creeping vine that roots in the ground and grows up trellises or other supporting frames, wrapping around supports with thin, spiraling tendrils.[7] The plant may also root in a soilless medium, whereby it will sprawl along the ground in lieu of a supporting structure. The vine has large leaves that form a canopy over the fruits. The fruit of typical cultivars of cucumber is roughly cylindrical, but elongated with tapered ends, and may be as large as 62 centimeters (24 in) long and 10 centimeters (4 in) in diameter. Cucumber fruits consist of 95% water (see nutrition table). In botanical terms, the cucumber is classified as a pepo, a type of botanical berry with a hard outer rind and no internal divisions. However, much like tomatoes and squashes, it is often perceived, prepared, and eaten as a vegetable"
summary_text=summary(paragraph,max_length=60,min_length=30,do_sample=False)
print(summary_text[0]['summary_text'])
```

```python
# Translation
translation=pipeline("translation_en_to_fr")
text="My name is Divya"
translation_text=translation(text,max_length=40)
print(translation_text[0]['translation_text'])
```

```python
# Sentiment Analysis
Sentiment=pipeline("sentiment-analysis")
text="Sun rises in the North"
sentiment_score=Sentiment(text)
print(sentiment_score)
```
